```python
import pandas as pd
from pymongo import MongoClient
from sklearn.preprocessing import StandardScaler

db = MongoClient('localhost', 27017)
cursor = db.sustaindb['county_stats'].find()
df = pd.DataFrame(list(cursor))
df
```

```python
features = [
    'total_population',
    'median_household_income',
    'no_of_hospitals',
    'no_of_electrical_stations',
    'no_of_fire_stations'
    'no_of_power_plants'
]

print(f'#features:{len(features)}')

x = df.loc[:, features].values

y = df.loc[:, 'RPL_THEMES'].values

x = StandardScaler().fit_transform(x)
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=34, random_state=2021)

# principalComponents = pca.fit_transform(x)
x_pca_34 = pca.fit_transform(x)

print(f'Variance explained by all 34 PCs = {pca.explained_variance_ratio_} * 100')
print(len(pca.explained_variance_ratio_))
```

```python
pcs = pd.DataFrame(pca.components_,columns=features, 
                   index = [f'pc_{x}' for x in range(1,len(features)+1)])
pcs
```
